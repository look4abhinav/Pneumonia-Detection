```python
import pandas as pd
import plotly as px
import numpy as np
pd.options.plotting.backend = "plotly"
```

```python
platinum = pd.read_csv("platinum_prices.csv", dtype_backend="pyarrow")
gold = pd.read_csv("gold_prices.csv", dtype_backend="pyarrow")
silver = pd.read_csv("silver_prices.csv", dtype_backend="pyarrow")
```

```python
platinum["Date"] = pd.to_datetime(platinum["Date"], utc=True)
gold["Date"] = pd.to_datetime(gold["Date"], utc=True)
silver["Date"] = pd.to_datetime(silver["Date"], utc=True)
```

```python
platinum.set_index("Date", drop=True, inplace=True)
gold.set_index("Date", drop=True, inplace=True)
silver.set_index("Date", drop=True, inplace=True)
```

```python
prices = pd.concat((platinum, gold, silver), axis="columns", join="inner")
prices.dropna(axis="index", how="any", inplace=True)
```

```python
prices[["platinum_kg_usd_price", "gold_kg_usd_price", "silver_kg_usd_price"]].plot()
```
